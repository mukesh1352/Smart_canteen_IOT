```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# Load JSON data
file_path = "pir-data-a8205-default-rtdb-export (5).json"
with open(file_path, "r") as file:
    data = json.load(file)

# Extract logs
event_logs = data.get("logs", {})

# Convert logs to DataFrame
records = []
for timestamp, details in event_logs.items():
    records.append({
        "timestamp": int(timestamp),
        "event": details.get("event", ""),
        "peopleCount": details.get("peopleCount", 0)
    })

df = pd.DataFrame(records)
df.sort_values(by="timestamp", inplace=True)

# Convert timestamp to datetime
df["timestamp"] = pd.to_datetime(df["timestamp"], unit='ms')

# Plot people count over time
plt.figure(figsize=(12, 6))
plt.plot(df["timestamp"], df["peopleCount"], marker='o', linestyle='-', color='b', label="People Count")
plt.xlabel("Time")
plt.ylabel("People Count")
plt.title("People Count Over Time")
plt.legend()
plt.xticks(rotation=45)
plt.grid()
plt.show()

# Summary statistics
print("\nSummary Statistics:")
print(df["peopleCount"].describe())

# Peak & Low occupancy times
peak_time = df[df["peopleCount"] == df["peopleCount"].max()]
low_time = df[df["peopleCount"] == df["peopleCount"].min()]

print("\nPeak Occupancy:")
print(peak_time)

print("\nLowest Occupancy:")
print(low_time)
```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import os

# List of JSON files to process
file_paths = [
    "water-depth-default-rtdb-export (4).json",
    "water-depth-default-rtdb-export (3).json",
    "water-depth-default-rtdb-export (2).json",
    "water-depth-default-rtdb-export (1).json",
    "water-depth-default-rtdb-export (5).json"
]

def process_json(file_path):
    if not os.path.exists(file_path):
        print(f"⚠ File not found: {file_path}")
        return None, file_path
    
    with open(file_path, "r") as file:
        data = json.load(file)
    
    # Extract water level logs
    water_logs = data.get("waterLog", {})
    
    # Convert logs to DataFrame
    records = []
    for timestamp, details in water_logs.items():
        if "level" in details:
            records.append({
                "timestamp": int(timestamp),
                "level": details.get("level", 0)
            })
    
    df = pd.DataFrame(records)
    if df.empty:
        print(f"⚠ No valid data in {file_path}")
        return None, file_path
    
    df.sort_values(by="timestamp", inplace=True)
    
    # Convert timestamp to datetime with error handling
    try:
        df["timestamp"] = pd.to_datetime(df["timestamp"], unit='ms', errors='coerce')
    except Exception as e:
        print(f"⚠ Timestamp conversion failed for {file_path}: {e}")
    
    return df, file_path

# Process each file and generate individual graphs
for file_path in file_paths:
    df, filename = process_json(file_path)
    if df is None:
        continue
    
    # Plot water level over time
    plt.figure(figsize=(12, 6))
    plt.plot(df["timestamp"], df["level"], marker='o', linestyle='-', label=f"Water Level - {filename}")
    plt.xlabel("Time")
    plt.ylabel("Water Level")
    plt.title(f"Water Level Over Time ({filename})")
    plt.legend()
    plt.xticks(rotation=45)
    plt.grid()
    plt.show()
    
    # Summary statistics
    print(f"\nSummary Statistics for {filename}:")
    print(df["level"].describe())
    
    # Peak & Low water levels
    peak_level = df[df["level"] == df["level"].max()]
    low_level = df[df["level"] == df["level"].min()]
    
    print("\nPeak Water Level:")
    print(peak_level)
    
    print("\nLowest Water Level:")
    print(low_level)
```
