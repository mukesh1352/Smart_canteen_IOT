```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# Load JSON data
file_path = "1.json"
with open(file_path, "r") as file:
    data = json.load(file)

# Extract logs
event_logs = data.get("logs", {})

# Convert logs to DataFrame
records = []
for timestamp, details in event_logs.items():
    records.append({
        "timestamp": int(timestamp),
        "event": details.get("event", ""),
        "peopleCount": details.get("peopleCount", 0)
    })

df = pd.DataFrame(records)
df.sort_values(by="timestamp", inplace=True)

# Convert timestamp to datetime
df["timestamp"] = pd.to_datetime(df["timestamp"], unit='ms')

# Plot people count over time
plt.figure(figsize=(12, 6))
plt.plot(df["timestamp"], df["peopleCount"], marker='o', linestyle='-', color='b', label="People Count")
plt.xlabel("Time")
plt.ylabel("People Count")
plt.title("People Count Over Time")
plt.legend()
plt.xticks(rotation=45)
plt.grid()
plt.show()

# Summary statistics
print("\nSummary Statistics:")
print(df["peopleCount"].describe())

# Peak & Low occupancy times
peak_time = df[df["peopleCount"] == df["peopleCount"].max()]
low_time = df[df["peopleCount"] == df["peopleCount"].min()]

print("\nPeak Occupancy:")
print(peak_time)

print("\nLowest Occupancy:")
print(low_time)
```
